```python
from PairTrading_eng import *
import pytz
from ib_insync import *
import datetime
import pandas_market_calendars as mcal
import yfinance as yf
```

```python
end_time = datetime.datetime.today()
start_time = end_time - pd.DateOffset(months = 26)
```

```python
import plotly.graph_objects as go

data_1d = yf.download('PARA', start=start_time, end=end_time)

# Assuming data_1d is already populated with the stock data
fig = go.Figure()

# Add trace for closing prices
fig.add_trace(go.Scatter(x=data_1d.index, y=data_1d['Close'], mode='lines', name='Close'))

# Customize layout
fig.update_layout(title='Stock Closing Prices', xaxis_title='Date', yaxis_title='Price')

# Show plot
fig.show()
```
